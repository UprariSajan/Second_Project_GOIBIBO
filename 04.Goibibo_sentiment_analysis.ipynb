```python
from textblob import TextBlob
import pandas as pd
```

```python
analysis=pd.read_excel('sentiment_analysis.xlsx')
```

```python
analysis
```

```python
polarity_rate=[]
for i in range(0,analysis.shape[0]):
    score=TextBlob(analysis.iloc[i][1])
    score1=score.sentiment[0]
    polarity_rate.append(score1)
```

```python
goibibo=pd.concat([analysis,pd.Series(polarity_rate)],axis=1)
```

```python
goibibo.tail()
```

```python
goibibo.rename(columns={goibibo.columns[2]:'Sentiment_analysis'},inplace=True)
```

```python
goibibo
```

Polarity tell us how Positive or Negative the comment is 
Value of Polarity is range of -1 to 1 (-ve value indicates sentiment is not good)
                                      (+ve value indicates sentiment is good)
