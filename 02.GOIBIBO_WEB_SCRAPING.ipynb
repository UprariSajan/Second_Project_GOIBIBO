```python
soup=BeautifulSoup(webpage,'lxml')   #passing our varibale in beautifulsoup
soup
```

```python
len(soup.find_all('h2'))   
```

```python
soup.find_all('div',class_="Layouts__Column-sc-1yzlivq-3 RnRstyles__UserReviewOuterWrapperDiv-sc-1i6bb0l-9 dSpsBx blA-DGS")
```

```python
lst1=[]
for i in (soup.find_all('div',class_='HotelCardstyles__HotelNameWrapperDiv-sc-1s80tyk-22 bwAcIn')):
    lst1.append(i.text.strip())

```

```python
lst2=[]
for i in (soup.find_all('div',type='customLocation')):
    lst2.append(i.text.strip())
```

```python
lst3=[]
for i in (soup.find_all('span',class_='ReviewAndRatingsstyles__AverageReviewText-sc-1nxmeoo-10 ZTpPO')):
    lst3.append(i.text.strip())
```

```python
lst4=[]
for i in (soup.find_all('span',itemprop="ratingCount")):
    lst4.append(i.text.strip())
```

```python
lst5=[]
for i in (soup.find_all('div',type="belowBannerPersuasion")):
   lst5.append(i.text.strip())
```

```python
lst6=[]
for i in (soup.find_all('p',class_="HotelCardstyles__OriginalPriceTextWrapper-sc-1s80tyk-38 dkyPLR")):
    lst6.append(i.text.strip())
```

```python
lst7=[]
for i in (soup.find_all('p',itemprop="priceRange")):
    lst7.append(i.text.strip())
```

```python
lst8=[]
for i in (soup.find_all('span',class_="HotelCardstyles__TaxPriceTextWrapper-sc-1s80tyk-41 brnSez")):
   lst8.append(i.text.strip())
```

```python
divs=soup.find_all('div',itemtype="http://schema.org/Hotel")
len(divs)
```

```python
df=pd.DataFrame(list(zip(lst1,lst2, lst3,lst4,lst5,lst6,lst7,lst8)),
              columns=['Hotel_Name','Location','Rating','Total_verified_Ratings','Cancilation_Fee','Price','Offer_Price','Taxes_Fees'])
```
